# 3. Convert to its feature of ids, attention mask in a vector format

```python
def convert_examplestofeatures(example_batch):
    # Filter out any None or non-string values
    inputs = [x if isinstance(x, str) else "" for x in example_batch['dialogue']]
    targets = [x if isinstance(x, str) else "" for x in example_batch['summary']]

    # Tokenize input dialogues
    input_encodings = tokenizer(
        inputs,
        max_length=1024,
        padding='max_length',
        truncation=True
    )

    # Tokenize target summaries
    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(
            targets,
            max_length=128,
            padding='max_length',
            truncation=True
        )

    return {
        'input_ids': input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

# 4. Perform the vectorization for all the dataset

```python
# The data set size is big so we give it as batches

dataset_samsum_pt=dataset_samsum.map(convert_examplestofeatures,batched=True)
```

```python
trainer=Trainer(model=model_pegasus,
                args=training_args,
                tokenizer=tokenizer,
                data_collator=seq2seq_data_collator,
                train_dataset=dataset_samsum_pt["test"], ## data set is large so dont use the train data
                eval_dataset=dataset_samsum_pt["validation"]
                )
```

### gen_kwargs = {"length_penalty": 0.8, "num_beams": 8, "max_length": 128}
1. length_penalty: 0.8
Adjusts the model's preference for shorter vs. longer sequences.

< 1.0 (like 0.8) = prefers shorter outputs

> 1.0 = encourages longer outputs

Useful to control overly verbose or overly brief generations.

2. num_beams: 8
Enables beam search with 8 beams.

Beam search keeps track of the top 8 most likely sequences at each step, rather than just 1 (greedy).

Higher value = better quality (more exhaustive), but slower.

Common values: 4–10.

3. max_length: 128
Maximum number of tokens the model can generate in one output sequence.

Helps avoid infinite loops or excessively long outputs.

Output will stop once this length is reached (or if EOS token is generated earlier).

```python
# prediction

gen_kwargs={"length_penalty":0.8,"num_beams":8,"max_length":128}

# dialogue from the test data for prediction
sample_text=dataset_samsum["test"][0]['dialogue']

# sample summary of the test dataset
actual_summary=dataset_samsum["test"][0]['summary']

pipe=pipeline("summarization",model="pegasus-samsum-model",tokenizer=tokenizer)

print("Acutual dialogue")
print(sample_text)

print("Actual Summary")
print(actual_summary)

print("Predicted Summary")
print(pipe(sample_text,**gen_kwargs)[0]['summary_text'])
```
