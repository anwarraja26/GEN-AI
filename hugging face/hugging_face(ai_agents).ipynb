```python
from transformers import AutoTokenizer

tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")

text="I am happy now "

tokens=tokenizer.tokenize(text)

print(f"Tokens: {tokens}" )
```

## convert tokens to vector

```python
input_ids=tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs: ",input_ids)
```

## Do the tokinization and the vector in same place using the method of encoding

```python
encoded_text=tokenizer(text)
print(encoded_text) # start of sentence is represent by 101 and end is represented by 102

```

## Decode the token id

```python
decode_output=tokenizer.decode(input_ids)
print(decode_output)
```

# Execute in a flow

```python
from transformers import AutoTokenizer

tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")

text="I am happy now "
print(text)
tokens=tokenizer.tokenize(text)

print(f"Tokens: {tokens}" )

input_ids=tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs: ",input_ids)

encoded_text=tokenizer(text)
print(encoded_text) # start of sentence is represent by 101 and end is represented by 102

decode_output=tokenizer.decode(input_ids)
print(decode_output)
```

# Fine Tunign IMDB

```python
from datasets import load_dataset
dataset = load_dataset('imdb')
```

```python
dataset
```

```python
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")
def tokenize_function(examples):
  return tokenizer(examples['text'],padding="max_length",truncation=True)

tokenized_datasets=dataset.map(tokenize_function,batched=True)
print(tokenized_datasets)
```

```python
!pip install -U transformers -q
```

## Specifyign the hyper parameter and arguments


```python
from transformers import TrainingArguments

training_args=TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=1,
    weight_decay=0.01,
    run_name="bert-seqclass-run-01",
)
training_args
```

```python
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(500))
```

```python
from transformers import AutoModelForSequenceClassification,Trainer
model=AutoModelForSequenceClassification.from_pretrained("bert-base-uncased",num_labels=2)

trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,

)
```

```python
# import os
# os.environ["WANDB_DISABLED"] = "true"
```

```python
trainer.train()
```

```python
results=trainer.evaluate()
print(results)
```

```python
model.save_pretrained('./fine-tuned-model')
tokenizer.save_pretrained('./fine-tuned-model')
```

```python
import arxiv
import pandas as pd

query="AI among student and professionals"
search=arxiv.Search(query=query,max_results=10,sort_by=arxiv.SortCriterion.SubmittedDate)

papers=[]
for result in search.results():
  papers.append({
      'published':result.published,
      'title':result.title,
      'abstract':result.summary,
      'categories':result.categories
  })

df=pd.DataFrame(papers)

df.head()
```

```python
df.shape
```

```python
abstract=df['abstract'][0]
summarizer=pipeline("summarization",model="facebook/bart-large-cnn")
result=summarizer(abstract)
```
